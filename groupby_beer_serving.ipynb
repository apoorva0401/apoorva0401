```python
import pandas as pd
import matplotlib as plt
```

```python
beer_df = pd.read_csv("http://bit.ly/drinksbycountry")
```

```python
beer_df
```

```python
# you can also use dot agg (aggregate functions) to explain
beer_df.groupby(['country']).agg({'beer_servings':['mean','max','sum','count'], 'total_litres_of_pure_alcohol':['mean','max','sum','count']})
```

```python
beer_df.groupby(['country']).beer_servings.mean().plot(kind='bar');
```

```python
plt.style.use('ggplot')
beer_df.groupby(['continent']).beer_servings.mean().plot(kind='bar');
```

```python
beer_df.groupby(['continent','country']).agg({'beer_servings':['mean','max','sum','count']})
```
